```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as py
```

```python
df = pd.read_csv("ToyotaCorolla.csv")
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
fuel_ty = pd.get_dummies(df['Fuel_Type'],drop_first=True)
fuel_ty[:5]
```

```python
df = pd.concat([df,fuel_ty],axis=1)
df.head()
```

```python
df.columns
```

```python
df.drop(['Id', 'Model', 'Age_08_04', 'Mfg_Month', 'Mfg_Year', 'KM',
       'Fuel_Type', 'HP', 'Met_Color', 'Color', 'Automatic', 'CC', 'Doors',
        'Quarterly_Tax', 'Weight' ],axis=1,inplace=True)
```

```python
df.head()
```

```python
X = df.drop(['Price'],axis=1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.6,random_state=101)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(random_state=101)
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
pred
```

```python
from sklearn.tree import plot_tree
py.figure(figsize=(10,8), dpi=150)
plot_tree(model, feature_names=X.columns)
```
